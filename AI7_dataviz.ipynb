```
count,bin_edges = np.histogram(dt,15)
dt.plot(kind="hist",
        figsize=(10,6),
        bins=15,
        alpha=0.6,
        xticks=bin_edges,
        color=['pink','yellow','blue'])
plt.title('Histogram of Immigration from Denmark, Norway, and Sweden from 1980 - 2013')
plt.ylabel('Number of Years')
plt.xlabel('Number of Immigrants')

plt.show()
```

```
count,bin_edges = np.histogram(dt,15)
x_min = bin_edges[0] - 10
x_max = bin_edges[-1] + 10
dt.plot(kind="hist",
        figsize=(10,6),
        bins=15,
        alpha=0.6,
        xticks=bin_edges,
        stacked=True,
        xlim=(x_min,x_max),
        color=['pink','yellow','blue'])
plt.title('Histogram of Immigration from Denmark, Norway, and Sweden from 1980 - 2013')
plt.ylabel('Number of Years')
plt.xlabel('Number of Immigrants')

plt.show()
```

```
data.sort_values(by="Total",ascending=True,inplace=True)
data15 = data["Total"].tail(15)
data15
```

```
data15.plot(kind="barh",figsize=(12,12),color='steelblue')
plt.xlabel("Number of Immigrants")
plt.title('Top 15 Conuntries Contributing to the Immigration to Canada between 1980 - 2013')
for index,value in enumerate(data15):
  label = format(int(value),',')
  plt.annotate(label,xy=(value - 47000,index - 0.10),color='white')
plt.show() 
```
