```python
import pandas as pd 
import matplotlib.pyplot as plt 
import numpy as np 
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster import hierarchy
from scipy.spatial import distance_matrix
```

```python
!wget -O cars_clus.csv https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/ML0101ENv3/labs/cars_clus.csv
```

```python
data = pd.read_csv("cars_clus.csv")
data.head()
```

```python
data.columns
```

```python
print("shape of the data before cleaning: ",data.shape)
data[['sales', 'resale', 'type', 'price', 'engine_s',
       'horsepow', 'wheelbas', 'width', 'length', 'curb_wgt', 'fuel_cap',
       'mpg', 'lnsales', 'partition']] = data[['sales', 'resale', 'type', 'price', 'engine_s',
       'horsepow', 'wheelbas', 'width', 'length', 'curb_wgt', 'fuel_cap',
       'mpg', 'lnsales', 'partition']].apply(pd.to_numeric,errors="coerce")
data = data.dropna()
data = data.reset_index(drop=True)
print("shape of the data after after cleaning: ",data.shape)
data.head()
```

```python
features = data[['engine_s',  'horsepow', 'wheelbas', 'width', 'length', 'curb_wgt', 'fuel_cap', 'mpg']]
```

```python
from sklearn.preprocessing import MinMaxScaler
x = features.values
nf = MinMaxScaler().fit_transform(x)
nf[0:5]
```

```python
nf.shape
```

```python
import scipy
leng = nf.shape[0]
d = scipy.zeros([leng,leng])
for i in range(leng):
  for j in range(leng):
    d[i,j] = scipy.spatial.distance.euclidean(nf[i],nf[j])
d.shape
```

```python
import pylab 
import scipy.cluster.hierarchy
z = hierarchy.linkage(d,"complete")
```

```python
from scipy.cluster.hierarchy import fcluster
k = 5
clusters = fcluster(z,k,criterion="maxclust")
clusters
```

```python
fig = pylab.figure(figsize=(18,50))
def llb(id):
  return '[%s,%s,%s]' % (data['manufact'][id],data['model'][id],int(float(data["type"][id])))
dendro = hierarchy.dendrogram(z,leaf_label_func=llb,leaf_rotation=0,leaf_font_size=12,orientation="right")
```

```python
dist_matrix = distance_matrix(nf,nf) 
print(dist_matrix)
```

```python
agglom = AgglomerativeClustering(n_clusters=6,linkage="complete")
agglom.fit(nf)
agglom.labels_
```

```python
data["clusters_"] = agglom.labels_
data.head()
```

```python
import matplotlib.cm as cm 
n_clusters = max(agglom.labels_) + 1
colors = cm.rainbow(np.linspace(0,1,n_clusters))
cluster_labels = list(range(0,n_clusters))
plt.figure(figsize=(16,14))
for color,label in zip(colors,cluster_labels):
  subset = data[data.clusters_ == label]
  for i in subset.index:
    plt.text(subset.horsepow[i],subset.mpg[i],str(subset['model'][i]),rotation=25)
  plt.scatter(subset.horsepow,subset.mpg,s=subset.price*10,c=color,label="cluster"+str(label),alpha=0.5)
plt.legend()
plt.xlabel("horsepow")
plt.ylabel("mpg")
plt.title("Clusters")
```

```python
data.groupby(["clusters_","type"])["clusters_"].count()
```

```python
agg_cars = data.groupby(['clusters_','type'])['horsepow','engine_s','mpg','price'].mean()
agg_cars
```

```python
plt.figure(figsize=(16,10))
for color, label in zip(colors, cluster_labels):
    subset = agg_cars.loc[(label,),]
    for i in subset.index:
        plt.text(subset.loc[i][0]+5, subset.loc[i][2], 'type='+str(int(i)) + ', price='+str(int(subset.loc[i][3]))+'k')
    plt.scatter(subset.horsepow, subset.mpg, s=subset.price*20, c=color, label='cluster'+str(label))
plt.legend()
plt.title('Clusters')
plt.xlabel('horsepow')
plt.ylabel('mpg')
```
